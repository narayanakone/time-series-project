```python
def ADF(data):
    result = adfuller(data)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]:.20f}')
    for key, value in result[4].items():
        print('Critial Values:', f'   {key}, {value}')
```

### Build Model

```python
train_data_first_diff = data_first_diff[:length_train]
test_data_first_diff  = data_first_diff[length_train:]
```

### AR

```python
model_ar = ARIMA(train_data_first_diff, order=(1,0,0))
model_fit_ar = model_ar.fit()
model_fit_ar.summary()
```

```python
pred_AR_diff = pd.Series(model_fit_ar.fittedvalues, copy=True)
print(pred_AR_diff.head())
```

```python
pred_AR = pd.Series(df['rgnp'], index=df['rgnp'].index)
pred_AR = pred_AR.add(pred_AR_diff, fill_value=0)

plt.plot(df['rgnp'], label = 'Actual',c='blue')
plt.plot(pred_AR, label = 'AR model',c='red')
plt.legend()
plt.title("AR model - Real Gross National Product ['rgnp']")
plt.show()
```

###  INFERENCE: <font color='blue'> The model includes one autoregressive term AR(1), no differencing (D=0), and no moving average term MA(0). The inclusion of the AR(1) term indicates that the current value of rgnp is influenced by its value from the previous period. The AR(1) coefficient is 0.2319 with a standard error of 0.101 and a p-value of 0.022. This positive coefficient indicates a moderate but statistically significant positive correlation with the previous time periods value. </font> 

# MA

```python
model_ma = ARIMA(train_data_first_diff, order=(0,0,2))
model_fit_ma = model_ma.fit()
model_fit_ma.summary()
```

```python
pred_MA_diff = pd.Series(model_fit_ma.fittedvalues, copy=True)
pred_MA = pd.Series(df['rgnp'], index=df['rgnp'].index)
pred_MA = pred_MA.add(pred_MA_diff, fill_value=0)

plt.plot(df['rgnp'], label = 'Actual',c='blue')
plt.plot(pred_MA, label = 'MA model',c='red')
plt.legend()
plt.title("MA model - Real Gross National Product ['rgnp']")
plt.show()
```

###  INFERENCE: <font color='blue'>  ARIMA(0,0,2) model on the rgnp data includes two moving average (MA) terms and no differencing (D=0) or autoregressive (AR) terms.It shows a good fit with a significant constant term (17.5392, p < 0.0001) and two moving average terms (MA(1) = 0.2092, p = 0.055; MA(2) = 0.1864, p = 0.046). The model has a lower AIC (928.257) compared to the ARIMA(1,0,0) model, indicating a slightly better fit. However, residual diagnostics indicate issues with non-normality (Jarque-Bera p = 0.03) and heteroskedasticity (H test p = 0.00).

```python
predict_varma = fitted_model.get_prediction()
predictions_varma=predict_varma.predicted_mean
predictions_varma.columns=['ulc_predicted','rgnp_predicted']
predictions_varma
```

```python
test_vs_pred=pd.concat([test_df,predictions_varma],axis=1)
test_vs_pred.plot(figsize=(12,5))
```

###  INFERENCE: <font color='blue'> The VARMA(4,3) model does not find significant coefficients for most lags of ulc and rgnp, suggesting that higher-order terms do not significantly contribute to the model.The residuals show no significant autocorrelation and are normally distributed, but heteroskedasticity is present, indicating variable variance over time.The significant negative covariance between the residuals of ulc and rgnp suggests an inverse relationship in their unexplained variations.The lack of significant coefficients may imply overfitting or an overly complex model. Simplifying the model, addressing heteroskedasticity, or including additional relevant variables might improve model performance.</font>
